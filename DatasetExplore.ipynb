Note that test set doesn't have data for the dependent variable - Sale Price. Hence, only train set can be used for the analysis.

```python
# correlation matrix
corr_matrix = df1.corr()
corr_matrix.style.background_gradient(cmap='coolwarm')  
```

```python
sorted = corr_matrix["SalePrice"].sort_values(ascending=False)
sorted
```

```python
# subplots for top 10 correlated numeric variables

fig, axes = plt.subplots(5,2, figsize=(10, 15))
axes = axes.flatten()
for i in range(len(sorted[1:11])): # exclude SalePrice
    sns.regplot(x=sorted.index[i+1], y='SalePrice', data=df1, ax=axes[i])
    axes[i].set_xlabel(sorted.index[i+1])
    axes[i].set_ylabel('SalePrice')
plt.tight_layout()
plt.show()
```
