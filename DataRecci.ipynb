### Filling missing values for nominal categories with the mode of the column


```python
df3.info()
```

```python
for i in nominal_categories:
    if i in df3.columns:
        df3[i].fillna(df3[i].mode(), inplace=True)
    
df3.info()
```

```python
for i in nominal_categories:
    if i in df3.columns:
        sns.catplot(x=i, y='SalePrice', data=df3, kind='box', height=5, aspect=2)
        plt.show()
```

```python
# calculate correlation measure between sale price and nominal variables
from scipy.stats import chi2_contingency
from scipy.stats import chi2

chi2_lis = []
for i in nominal_categories:
    if i in df3.columns:
        chi2 = chi2_contingency(pd.crosstab(df3[i], df3['SalePrice']))
        chi2_lis.append((i, ('statistic', chi2[0]), ('pvalue',chi2[1])))

# sort chi2 list based on p-value
chi2_lis.sort(key=lambda x: x[2][1])
chi2_lis

# WARNING: This may not make sense because chi 2 contingency test is not appropriate for continuous variables
```

```python
chi2
```

```python
# Calculate kruskal test for sale price and nominal variables
from scipy.stats import kruskal

kruskal_lis = []
for i in nominal_categories:
    if i in df3.columns:
        #group by nominal variable
        grouped = df3.groupby(i)
        # create list of grouped sale prices
        grouped_lis = []
        for j in grouped:
            grouped_lis.append(j[1]['SalePrice'])
        # kruskal test
        kruskal_temp = kruskal(*grouped_lis)
        kruskal_lis.append((i, ('statistic', kruskal_temp.statistic), ('pvalue',kruskal_temp.pvalue)))

# sort kruskal list based on p-value
kruskal_lis.sort(key=lambda x: x[2][1])
kruskal_lis
```

Based on the Kruskal-Wallis H-test, the following columns are statistically insignificant and hence can be dropped from the analysis:

- 'Condition2' (p-value = 17.77)
